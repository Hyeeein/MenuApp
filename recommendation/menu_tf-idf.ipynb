```python
import numpy as np
import pandas as pd
import sys
from sklearn.feature_extraction import text
from sklearn.metrics.pairwise import cosine_similarity
```

```python
menu = pd.read_csv('menu.csv')
menu.head()
```

```python
nutrient = pd.read_csv('nutrient.csv')
```

```python
# nutrient_menu 데이터의 ingredient 피처를 menu_id 기준으로 정렬
nutrient = nutrient.sort_values(by=['menu_id'])
nutrient.head()
```

```python
tmp_df = nutrient['ingredient']
tmp_df = tmp_df.reset_index(drop=True)

# 메뉴 데이터 특징 피처 만들어서 저장
menu['feature'] = menu['category'] + " " + menu['name'] + " " + menu['weather'] + " " + menu['emotion'] + " " + tmp_df
menu_feature = menu['feature'].tolist()
menu_feature
```

```python
len(menu_feature)
```

```python
menu_feature[3]
```

```python
# TF-IDF Vectorizer Object 구현
tfidf = text.TfidfVectorizer()
tfidf_features = tfidf.fit_transform(menu_feature)

tfidf_features = np.array(tfidf_features.todense())
np.set_printoptions(threshold=sys.maxsize)
tfidf_features
```

```python
# 코사인 유사도 계산
similarity = cosine_similarity(tfidf_features)
```

```python
# 중복된 추천 제외하고, 특징 데이터 사용
indices = pd.Series(menu.index, index=menu.feature).drop_duplicates()
```

```python
# 예: 4번째 얼큰순대국과 유사한 메뉴 추천
feature = menu_feature[3]

index = indices[feature]
similarity_scores = list(enumerate(similarity[index]))
similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
similarity_scores = similarity_scores[1:11]
menu_indices = [i[0] for i in similarity_scores]
```
